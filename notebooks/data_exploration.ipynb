## Load to SQLite


```python
import sqlite3

# Create database
conn = sqlite3.connect('../database/retail.db')

# Load data into SQLite database
customers.to_sql('customers', conn, if_exists='replace', index=False)
products.to_sql('products', conn, if_exists='replace', index=False)
orders.to_sql('orders', conn, if_exists='replace', index=False)

# Check success
pd.read_sql("SELECT * FROM customers LIMIT 5", conn)

```

# Business Questions

## Total number of customers
Result: 1200 customers

```python

pd.read_sql("SELECT COUNT(customer_id) as count_of_customers FROM customers", conn)
```

## Total numbers of orders
Result: 4000 orders

```python
pd.read_sql("SELECT COUNT(order_id) as count_of_orders FROM orders", conn)
```

## Top cities by number of customers
Result: Sheffield, Nottingham, Birmingham, Bristol

```python
pd.read_sql(
    """SELECT city, COUNT(customer_id)
    FROM customers
    GROUP BY city
    ORDER BY COUNT(customer_id) DESC
    LIMIT 4
            """, conn)
```

## Total revenue
Result: 1644817.71

```python
# JOIN product_id from different tables
pd.read_sql("""
SELECT SUM(products.price * orders.quantity) as total_revenue
FROM orders
JOIN
products ON orders.product_id = products.product_id""", conn).head()
```

## Top 5 most purchased products
### Result:
1. Monitor 55 (201)
2. Wireless Mouse 60 (193)
3. Bluetooth Speaker 18 (191)
4. Water Bottle 71 (178)
5. USB Cable 65 (178)

```python
pd.read_sql("""
SELECT SUM(orders.quantity),
    products.product_name
FROM orders
JOIN products ON orders.product_id = products.product_id
GROUP BY products.product_name
ORDER BY SUM(orders.quantity) DESC
""", conn).head()
```

## Top category by revenue
### Result: Home (473002.14)

```python
pd.read_sql("""
SELECT products.category as ranking_category_by_revenue,
    SUM(orders.quantity * products.price) as total_revenue_by_category
FROM orders
JOIN products ON orders.product_id = products.product_id
GROUP BY products.category
ORDER BY SUM(orders.quantity * products.price) DESC
""", conn).head()
```

## Average quantity per order
### Result: 2.97

```python
pd.read_sql("""
SELECT SUM(orders.quantity) * 1.0 / COUNT(DISTINCT orders.order_id) as average_order_quantity
FROM orders
""", conn).head()
```

## Top cities by revenue
### Result: Sheffield (233607.88)

```python
pd.read_sql("""
SELECT customers.city,
SUM(products.price * orders.quantity) as total_revenue_by_city
FROM orders
JOIN products ON orders.product_id = products.product_id
JOIN customers ON orders.customer_id = customers.customer_id
GROUP BY customers.city
ORDER BY SUM(products.price * orders.quantity) DESC
""", conn)
```

## Top 10 customers by spending

```python
pd.read_sql("""
SELECT orders.customer_id AS customer_id,
SUM(orders.quantity * products.price) as total_spending
FROM orders
JOIN products ON orders.product_id = products.product_id
GROUP BY orders.customer_id
ORDER BY SUM(orders.quantity * products.price) DESC
""", conn).head(10)
```

## Best month by revenue

### Result: top month by revenue is January of 2024 (127940.85)

```python
pd.read_sql("""
SELECT SUM(orders.quantity * products.price) as monthly_revenue,
strftime('%Y-%m', order_date) as month
FROM orders
JOIN products ON orders.product_id = products.product_id
GROUP BY strftime('%Y-%m', order_date)
ORDER BY monthly_revenue DESC
""",conn)
```

## Most common payment method
### Result: Cash (1362)

```python
pd.read_sql("""
SELECT COUNT(orders.payment_method) as payment_count,
    orders.payment_method
FROM orders
GROUP BY orders.payment_method
ORDER BY payment_count DESC
""", conn)
```
